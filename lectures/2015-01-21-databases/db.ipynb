```python
import pandas as pd
import sqlite3
```

```python
import os
try:
    os.remove("msdss.db")
except OSError:
    pass
```

```python
con = sqlite3.connect("msdss.db")
```

```python
con.execute("""
CREATE TABLE `photo` (
	`run`	INTEGER,
	`rerun`	INTEGER,
	`camcol`	INTEGER,
	`field`	INTEGER,
	`obj`	INTEGER,
	`type`	INTEGER,

	`ra`	REAL,
	`dec`	REAL,
	`psfMag_r`	REAL,
	`psfMag_g`	REAL,
	`psfMagErr_r`	REAL,
	`psfMagErr_g`	REAL
);
""")
```

```python
con.execute("""
CREATE TABLE `runs` (
	`run`	INTEGER,
	`ra`	REAL,
	`dec`	REAL,
	`mjdstart`	REAL,
	`mjdend`	REAL,
	`node`	REAL,
	`inclination`	REAL,
	`mu0`	REAL,
	`nu0`	REAL
);
""")
```

```python
runs = pd.read_csv('runs.txt', 
            sep=" ", header=None, skiprows=1, 
            names=['run', 'ra', 'dec', 'mjdstart', 'mjdend', 'node', 'inclination', 'mu0', 'nu0'],
            index_col = 'run')
```

```python
photo = pd.read_csv('sample.csv', index_col=['run', 'rerun', 'camcol', 'field', 'obj'])
```

```python
#con.execute("drop table if exists runs2")
runs.to_sql('runs', con, if_exists='append')
```

```python
#con.execute("drop table if exists photo2")
photo.to_sql('photo', con, if_exists='append')
```

```python
con.close()
del runs
del photo
```

```python
con = sqlite3.connect("msdss.db")
```

```python
result = pd.read_sql("""
    SELECT
        photo.ra, photo.dec, photo.run, mjdstart
    FROM
        photo
    JOIN runs ON photo.run = runs.run
""", con)
```
